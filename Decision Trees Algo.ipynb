```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
x = iris.data[:,:2]  # select all rows and first two features only
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.2,random_state=12)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()

dt.fit(xtrain,ytrain)

test_score = dt.score(xtest,ytest)

print('test score',test_score)

train_score = dt.score(xtrain,ytrain)

print('train score',train_score)
```

```python
dt.feature_importances_   ## Gives the score acc to feature's importance
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
cn =['setosa','versicolor','virginica']

plt.figure(figsize=(16,12))
tree.plot_tree(dt,feature_names = iris.feature_names[:2], class_names=cn, filled = True)

plt.show()
#plt.savefig('demo.png')
```

### Note:

- Test accuracy is low than train accuracy
- Overfitting problem when we have only first two features
- It means first two features are less important or less informative as compared to last two 

```python
x = iris.data[:,2:]  # select all rows and last two features only
y = iris.target
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.2,random_state=12)
```

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain,ytrain)

test_score = dt.score(xtest,ytest)

print('test score',test_score)

train_score = dt.score(xtrain,ytrain)

print('train score',train_score)
```

### Note:

- Better performance of model than last one
